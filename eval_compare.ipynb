```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data_c_index = {
    'CPH': cph_c_index,
    'Deepsurv': deepsurv_c_index,
    'Survival SVM': svm_c_index,
    'RSF': rsf_c_index
}

data_ibs = {
    'CPH': cph_ibs,
    'Deepsurv': deepsurv_ibs,
    'Survival SVM': [],
    'RSF': rsf_ibs
}

# Calculate mean values for each dataset, filtering out empty values
mean_values_c_index = {model: np.mean(values) for model, values in data_c_index.items() if values}
mean_values_ibs = {model: np.mean(values) for model, values in data_ibs.items() if values}

# Determine the dataset with the highest mean value for C-Index and the lowest mean value for IBS
max_mean_c_index = max(mean_values_c_index, key=mean_values_c_index.get)
min_mean_ibs = min(mean_values_ibs, key=mean_values_ibs.get)

# Plotting
plt.figure(figsize=(8, 5))

# Plot C-Index boxplot
bp_c_index = plt.boxplot(data_c_index.values(), patch_artist=True, showmeans=True,
            boxprops=dict(facecolor="white", color="black"),
            capprops=dict(color="black"),
            whiskerprops=dict(color="black"),
            flierprops=dict(color="black", markeredgecolor="black"),
            medianprops=dict(color="black"),
            meanprops=dict(marker='o', markerfacecolor="black", markeredgecolor='black', markersize=10)
            )

# Plot IBS boxplot
bp_ibs = plt.boxplot(data_ibs.values(), patch_artist=True, showmeans=True,
            boxprops=dict(facecolor="white", color="black"),
            capprops=dict(color="black"),
            whiskerprops=dict(color="black"),
            flierprops=dict(color="black", markeredgecolor="black"),
            medianprops=dict(color="black"),
            meanprops=dict(marker='o', markerfacecolor="black", markeredgecolor='black', markersize=10)
            )

# Customize x-axis ticks
plt.xticks(range(1, len(data_c_index) + 1), data_c_index.keys(), fontsize=12)

# Customize y-axis ticks and labels
plt.yticks(fontsize=12)
plt.ylabel('Values', fontsize=12)

# Set plot title
plt.title('Boxplot of C-Index & IBS for Different Models over 20 Replicates', fontsize=14)

# Highlight the best scoring model for C-Index with a red marker
bp_c_index['means'][list(data_c_index.keys()).index(max_mean_c_index)].set_markerfacecolor('red')

# Highlight the best scoring model for IBS with a blue marker
bp_ibs['means'][list(data_ibs.keys()).index(min_mean_ibs)].set_markerfacecolor('blue')

# Manually create legend with specified colors
legend_labels = ['C-Index', 'IBS', f'Highest C-Index: {max_mean_c_index}', f'Lowest IBS: {min_mean_ibs}']
legend_handles = [
    plt.Line2D([0], [0], color="black", linewidth=4),
    plt.Line2D([0], [0], color="black", linewidth=4),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markeredgecolor='black', markersize=10),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markeredgecolor='black', markersize=10)
]

# Add legend to the plot
plt.legend(legend_handles, legend_labels, loc='best', fontsize=12)

# Display the plot
plt.show()
```
